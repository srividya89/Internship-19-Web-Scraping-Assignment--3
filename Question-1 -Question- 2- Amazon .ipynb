```python
urls=[]
for page in range(0,3):
    
    product_url = driver.find_elements_by_xpath("//a[@class='a-link-normal a-text-normal']")
    for i in product_url:
        urls.append(i.get_attribute('href'))
    
    nxt_button=driver.find_elements_by_xpath("//a[@class='s-pagination-item s-pagination-next s-pagination-button s-pagination-separator']")
    try:                                                        
        driver.get(nxt_button[1].get_attribute('href'))
    except:
        driver.get(nxt_button[0].get_attribute('href'))
        
time.sleep(2) 
```

```python
urls
```

```python
product_names=[]
for i in urls:
    driver.get(i)
    desc=driver.find_elements_by_xpath("//span[@class='a-size-large product-title-word-break']")#scraping description from the xpath
    for i in desc:
        product_names.append(i.text)#appending the description in list
time.sleep(2)        
time.sleep(3)  
```

```python
product_names
```

```python
Brand_names=[]
for j in urls:
    driver.get(j)
    try:
        Brand_name=driver.find_element_by_xpath("//tr[@class='a-spacing-small']//td[2]//span[1]")
        Brand_names.append( Brand_name.text)
    
    except NoSuchElementException  as e:
        Brand_names.append("_")
            
Brand_names     
     
```

```python
Ratings=[]
for k in urls:
    driver.get(k)
    try:
        Rating=driver.find_element_by_xpath("//span[@data-hook='rating-out-of-text']")
        Ratings.append(Rating.text)
    
    except NoSuchElementException  as e:
        Ratings.append("_")
            
Ratings    
```

```python
No_of_Ratings=[]
for l in urls:
    driver.get(l)
    try:
        No_of_Rating=driver.find_element_by_xpath("//div[@data-hook='total-review-count']//span")
        No_of_Ratings.append(No_of_Rating.text)
    
    except NoSuchElementException  as e:
        No_of_Ratings.append("_")
            
No_of_Ratings    
```

```python
prices=[]
for l in urls:
    driver.get(l)
    try:
        price=driver.find_element_by_xpath("//span[@class='a-size-medium a-color-price priceBlockBuyingPriceString']")
        prices.append(price.text)
    
    except NoSuchElementException  as e:
        prices.append("_")
            
prices
```

```python
ret_exc=[]
for l in urls:
    driver.get(l)
    try:
        return_exc=''
        return_exc=driver.find_element_by_xpath("//div[@id='RETURNS_POLICY']")
        try:
             return_exc.click()
        except:
             time.sleep(1)
        time.sleep(1)
        return_exchange='' 
        return_exchange=driver.find_element_by_xpath("//h5[@class='secondary-header a-text-normal']").text
        ret_exc.append(return_exchange)
        
    except ElementClickInterceptedException  as e:
        time.sleep(1)    
            
    except NoSuchElementException  as e:
        time.sleep(1)
        ret_exc.append('_')
```

```python
ret_exc
```

```python
ret_exchange=[]
for l in urls:
    driver.get(l)
    try:
        return_exchange=''
        return_exchange=driver.find_element_by_xpath("//div[@id='RETURNS_POLICY']")
        try:
            return_exchange.click()
        except:
            time.sleep(1)
        return_exchangee='' 
        return_exchangee=driver.find_element_by_xpath("//div[@class='a-section policy']//span").text
        ret_exchange.append(return_exchangee)
        
    except ElementClickInterceptedException  as e:
        time.sleep(1)    
            
    except NoSuchElementException  as e:
        time.sleep(1)
        ret_exchange.append('_')
```

```python
ret_exchange
```

```python
expected_f_delivery=[]
for j in urls:
    driver.get(j)
    try:
        expected_f_deliveri=driver.find_element_by_xpath("//div[@id='ddmDeliveryMessage']//b")
        #for a in expected_deliveri:
        expected_f_delivery.append(expected_f_deliveri.text)
    
    except NoSuchElementException  as e:
        expected_f_delivery.append("_")
            
expected_f_delivery  
```

```python
expected_c_fast_delivery=[]
for j in urls:
    driver.get(j)
    try:
        expected_c_fast_deliveri=driver.find_element_by_xpath("//span[@id='upsell-message']")
        #for a in expected_deliveri:
        expected_c_fast_delivery.append(expected_c_fast_deliveri.text)
    
    except NoSuchElementException  as e:
        expected_c_fast_delivery.append("_")
            
expected_c_fast_delivery
```

```python
Availability=[]
for j in urls:
    driver.get(j)
    try:
        Availabiliti=driver.find_element_by_xpath("//span[@class='a-size-medium a-color-success']")
        #for a in expected_deliveri:
        Availability.append(Availabiliti.text)
    
    except NoSuchElementException  as e:
        Availability.append("_")
            
Availability
```

```python
Other_details_important=[]
for j in urls:
    driver.get(j)
    try:
        Other_details_importanti=driver.find_element_by_xpath("//table[@class='a-normal a-spacing-micro']//tr[7]")
        #for a in expected_deliveri:
        Other_details_important.append(Other_details_importanti.text)
    
    except NoSuchElementException  as e:
        Other_details_important.append("_")
            
Other_details_important
```
