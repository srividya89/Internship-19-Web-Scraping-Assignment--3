```python
urlss=[]
for page in range(0,3):#for loop for scrapping 4 page
    name_urls=driver.find_elements_by_xpath("//h2[@class='title title-6']//a")
    for i in name_urls:
        urlss.append(i.get_attribute('href'))
    try:
        nxt_button=driver.find_element_by_xpath("//div[@class='pagination-item pagination-last']")
        nxt_button.click()    
    except:
        time.sleep(1)        
```

```python
descriptions=[]
for j in urlss:
    driver.get(j)
    try:
        description=driver.find_element_by_xpath("//div[@class='content collapse-content']")
        descriptions.append(description.text)
    
    except NoSuchElementException  as e:
        descriptions.append("_")
            
descriptions    
     
```
