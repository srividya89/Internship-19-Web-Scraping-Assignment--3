```python
urls=[]
Product_urls=driver.find_elements_by_xpath("//a[@class='_1fQZEK']")
for i in Product_urls:
    urls.append(i.get_attribute("href"))
urls   
```

```python
Brand_names=[]
for j in urls:
    driver.get(j)
    try:
        Brand_name=driver.find_element_by_xpath("//div[@class='_1MR4o5']//div[4]//a")
        Brand_names.append( Brand_name.text)
    
    except NoSuchElementException  as e:
        Brand_names.append("_")
            
Brand_names    
```

```python
product_names=[]
for j in urls:
    driver.get(j)
    try:
        product_name=driver.find_element_by_xpath("//span[@class='B_NuCI']")
        product_names.append(product_name.text)
    
    except NoSuchElementException  as e:
        product_names.append("_")
            
product_names   
```

```python
clrs=[]

for j in urls:
    driver.get(j)
    try:
        clr=driver.find_elements_by_xpath("//a[@class='kmlXmn PP89tw']")
        for i in clr:
            clrs.append(i.get_attribute('href'))
            
    
    except NoSuchElementException  as e:
        clrs.append("_")
clrs
```
